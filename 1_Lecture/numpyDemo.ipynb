```python
#Simple Arithmetic Operation
x=np.array([10,20,30])
y=np.array([1,2,3])
print("X + Y:", x+y)
print("X - Y:", x-y)
print("X * Y:", x*y)
print("X / Y:", x//y)
```

```python
add = np.add(x,y) #addition function
print("addition", add)

#direct print
print("Substraction:", np.subtract(x,y)) #substract Function
print("Multiplication", np.multiply(x,y)) #multiply Function
print("Division", np.divide(x,y)) #divide Function
print("Module", np.mod(x,y)) #Module Function
```

```python
a= np.array([10,20,30,40,50])
print("Maximum value: ", np.max(a)) #maximum Function
print("Minimum value: ", np.min(a)) #minimum Function
print("total sum of value: ", np.sum(a)) #sum Function
print("Average value: ", np.mean(a)) #Average mean Function
```
