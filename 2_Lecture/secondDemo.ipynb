```python
import numpy as np
array = np.array([[25,64],[45,23]])
print(np.sqrt(array)) # Square function
```

```python
print(array)
print("\nrows become Column:\n", array.T)  #rows become column and column become rows
```

```python
b1 = np.array([[2,3],[4,5]])
b2 = np.array([[5,6],[7,8]])
#print(b1,"\n", b2)
matrix_multi = np.dot(b1, b2) #matrix multiplication function
print(matrix_multi)
```

```python
c1 = np.zeros([3,2]) # zeros Function
print(c1)
c1 = np.ones([3,2]) #ones function
print(c1, "\n",np.zeros(3))
```

```python
bool = np.array([[True, True],[True, False]])
print("all() with axis 0:", np.all(bool))  #return true false value
print("all() with axis none:", np.all(bool,axis=0))
```

```python
#arange(start,stop,step)
a1 = np.arange(1, 10, 2)
print(a1)
a2 = np.arange(9)
print(a2)

#reshape matrix
reshape = a2.reshape(3,3)
print("Reshape:\n", reshape)

#third colum remove
result = reshape[:, :2]
print("Only two column Reshape:\n",result)

result = reshape[:2, :]
print("Only two Row Reshape:\n",result)
```

```python
#vertically printing
x = np.array([[12,23],[23,43]])
y = np.array([[56,67],[78,89]])
z = np.array([[32,43],[54,65]])
print("X:\n",x,"\nY:\n", y, "\nZ:\n", z)

v= np.vstack((x,y,z)) #vstack combine vertically
print("Vertically vstack:\n", v)

h= np.hstack((x,y,z)) #hstack combine horizontally
print("Horizontally vstack:\n", h)
```

```python
# Vertical and horizontal split
arrr = np.array([[1,2,3,4],[5,6,7,8]])
print(arrr)

vsplit = np.vsplit(arrr, 2)
print("\nVertically split\n", vsplit)

hsplit = np.hsplit(arrr, 2)
print("\nHorizontally split\n", hsplit)
```

```python
jp = np.array([2,36,49,64])
print("Sqrt : ",np.sqrt(jp)) #squrt function
print("EXP : ", np.exp(jp)) #exponential function
print("sin : ", np.sin(jp)) #sin function
print("Cos : ", np.cos(jp)) #cos function
print("std : ", np.std(jp)) #standard division function
```
