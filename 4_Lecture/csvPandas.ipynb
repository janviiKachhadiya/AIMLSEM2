```python
import pandas as pd
stMarks = pd.read_csv('studentData.csv')
print(stMarks)
```

```python
# column value get
new_std = stMarks.dropna()
print(new_std)
```

```python
# nun value replace with particular value
fill_std = stMarks.fillna(111)
print(fill_std)
```

```python
#total marks
std1 = stMarks["Total Marks"].fillna(500)
#print(std1)
stMarks["Total Marks"].fillna(555,inplace = True)
print(stMarks)
```

```python
#mean,medium,mode
mean = stMarks["Obtained marks"].mean()
print(mean)
median = stMarks["Obtained marks"].median()
print(median)
mode = stMarks["Obtained marks"].mode()
print(mode)
```

```python
# Calculate the percentage
stMarks['Percentage'] = (stMarks['Obtained marks'] / stMarks['Total Marks']) * 100
print(stMarks)
```

```python
# Step 3: Apply the function to create a 'Grade' column
stMarks['Grade'] = stMarks['Percentage'].apply(
    lambda p: "A" if 80 <= p <= 100 else
              "B" if 60 <= p <= 80 else
              "C" if 40 <= p <= 60 else
              "D"
)

# Print the result
print(stMarks.head(5))
```
